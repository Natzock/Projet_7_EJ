```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../data/tableau_features.csv")
```

```python
df.shape
```

```python
# Number of each type of column
df.dtypes.value_counts()
```

```python
# Number of unique classes in each object column
df.select_dtypes('object').apply(pd.Series.unique, axis = 0)
```

```python
# Sélectionner les colonnes de type 'object'
object_columns = df.select_dtypes(include=['object']).columns

# Supprimer les colonnes de type 'object'
df = df.drop(columns=object_columns)
```

```python
# Identifier les colonnes contenant des valeurs infinies
inf_columns = df.columns[np.isinf(df).any()].tolist()
```

```python
inf_columns
```

```python
df[inf_columns]
```

```python
df.drop(columns=inf_columns, inplace=True)
```

```python
# Number of each type of column
df.dtypes.value_counts()
```
